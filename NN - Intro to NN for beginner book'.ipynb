```python
import numpy as np
w1 = np.array([[0.2, 0.2, 0.2], [0.4, 0.4, 0.4], [0.6, 0.6, 0.6]])
w2 = np.zeros((1, 3))
w2[0, :] = np.array([0.5,0.5, 0.5])
b1 = np.array([0.8, 0.8, 0.8])
b2 = np.array([0.2])
def f(x):
    return 1 / (1 + np.exp(-x)) 
```

### Matrix row to diagram
<img src="./matrix-to-diagram2.png" alt="Drawing" style="width: 400px;"/>

```python
w = [w1, w2]
b = [b1, b2]
x = [1.5, 2.0, 3.0]
w[1].shape
```

```python
def simple_loop_nn_calc(n_layers, x, w, b):
    for l in range(n_layers-1):
        if l == 0:
            node_in = x
        else:
            node_in = h
        h = np.zeros((w[l].shape[0], ))
        for i in range(w[l].shape[0]): #row in matrix
            f_sum = 0
            for j in range(w[l].shape[1]): #col in matrix
                f_sum += w[l][i][j] * node_in[j]
            print(l, f_sum)
            f_sum += b[l][i]
            h[i] = f(f_sum) #i is column
            print('f_sum : {0:.2f}'.format(f_sum))
    return h
```

```python
simple_loop_nn_calc(3, x, w, b)
```

```python
%timeit simple_loop_nn_calc(3, x, w, b)
```

### Matrix Multiplication
<img src="./7-4-2019-matrix-theory-1.png" alt="Drawing" style="width: 600px;"/>
<img src="./7-4-2019-matrix-theory-2.png" alt="Drawing" style="width: 600px;"/>

```python
def simple_loop_nn_calc_improve(n_layers, x, w, b):
    for l in range(n_layers-1):
        if l == 0:
            node_in = x
        else:
            node_in = h
        h = np.zeros((w[l].shape[0], ))
        z = w[l].dot(node_in) + b[l]
        print('z : {0}'.format(z))
        print('z.shape: {0}'.format(z.shape))
        h = f(z)
    return h
```

```python
w = [w1, w2]
b = [b1, b2]
x = [1.5, 2.0, 3.0]
print(w[1].shape)
simple_loop_nn_calc_improve(3, x, w, b)
```

### Gradient Descend

```python
x_old = 0
x_new = 6
gamma = 0.01 # step size
precision = 0.00001

def df(x):
    y = 4 * x**3 - 9 * x**2
    return y

while abs(x_new - x_old) > precision:
    x_old = x_new
    x_new += -gamma * df(x_old)
    print('x_old : {0:.4f}'.format(x_old))
    
print('The local minimum occurs at {0:.4f}'.format(x_new))
```
